```python
import pandas as pd 
df = pd.read_excel("Food expenditure in the US .xlsx")
print(df.head())
```

```python
print(df.describe())
```

```python
print(df.columns.tolist())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))

plt.plot(
    df["Year"],
    df["% of household disposable income spent on food at home (using current dollars)."],
    marker='o'
)

plt.title("% of household income spent on food at home (20152024)")
plt.xlabel("Year")
plt.ylabel("Percentage of income (%)")

plt.xticks(df["Year"])  # set ticks exactly to your years
plt.grid(True, linestyle='--', alpha=0.6)  # optional: adds readability

plt.show()
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(8,5))

plt.plot(df["Year"],df["% of household disposable income spent on food away from home (restaurants, cafes, etc.)."],marker = 'o')
plt.title("% of household income spent eating out (2015-2024)")
plt.xlabel("Year")
plt.ylabel("Percentage of income (%)")
plt.xticks(df["Year"])
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize= (8,5))
sns.barplot(x="Year", y="Total nominal food expenditure share of disposable personal income, household final users (percentage)",data = df,palette="Purples")
plt.title("% of household spent on food (at home + eat out) from 2010-2024")
plt.xlabel("Year")
plt.ylabel("Total spent (%)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,4))

df_melted = df.melt(
    id_vars="Year",
    value_vars=[
        "Out of all food spending, what % goes to food-at-home.",
        "Out of all food spending, what % goes to food-away."
    ],
    var_name="Category",
    value_name="Percentage"
)
sns.barplot(
    x="Year",
    y="Percentage",
    hue="Category",
    data=df_melted,
    palette=["purple","orange"]
)
plt.title("Food spending split: Home vs Away %")
plt.xlabel("Percentage %")
plt.ylabel("Year")
plt.xticks(rotation=45)
plt.legend(title="Category",bbox_to_anchor=(1,1),loc="upper left")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,4))
plt.plot(df["Year"], df["Average yearly spending per household on food at home (in current dollars)."],marker='o',label="Food-at-home")
plt.plot(df["Year"],df["Average yearly spending per household on food away from home (current dollars)."],marker='o',label="Eat-out")

plt.title("Average Yearly Food Spending (2015-2024)")
plt.xlabel("Year")
plt.ylabel("Dollars (USD)")
plt.legend(title="Category")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,4))
plt.plot(df["Year"], df["Total yearly household food spending (current dollars)."],marker='o',color="darkgreen")

plt.title("Total spending on food yearly (2015-2024)")
plt.xlabel("Year")
plt.ylabel("Dollars (USD)")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
